튜닝 파라미터의 모든 조합을 리턴하는 함수를 생성해보자.

```python
params = {'n_layer':[1, 2, 3],
          'units': [8, 16, 32],
          'optim': ['a', 'b', 'c']}
```

```python
keys = params.keys()
keys
```

```python
params['n_layer']
```

```python
retall = []
keys = params.keys()
for i, k in enumerate(keys):
    if i == 0:
        for v in params[k]:
            d = {}
            d[k] = v
            retall.append(d)
    else:
        tmp = retall.copy()
        retall = []
        for d in tmp:
            for v in params[k]:
                d = d.copy()
                d[k] = v
                retall.append(d)

retall
```

```python
def retallparams(params):
    retall = []
    keys = params.keys()
    for i, k in enumerate(keys):
        if i == 0:
            for v in params[k]:
                d = {}
                d[k] = v
                retall.append(d)
        else:
            tmp = retall.copy()
            retall = []
            for d in tmp:
                for v in params[k]:
                    d = d.copy()
                    d[k] = v
                    retall.append(d)

    return retall
```

```python
retallparams(params)
```

```python
import numpy as np
```

```python
h = np.array([-.5, 2.1, 0.3, -1.2])
h
```

```python
def relu(x):
    return np.maximum(x, 0)
```

```python
relu(h)
```

```python
def softmax(s):
    return np.exp(s) / np.exp(s).sum()
```

```python
prob = softmax(h)
prob
```

```python
prob.sum()
```

```python
prob.argmax()
```
